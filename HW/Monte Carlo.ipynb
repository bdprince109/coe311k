```python
import random
import math
def randomShots(numSamples):
    countTotal=0
    countCircle=0
    for i in range(0,numSamples):
        x=random.uniform(-1,1)
        y=random.uniform(-1,1)
        countTotal=countTotal+1
        if (math.sqrt(x*x+y*y)<1):
            countCircle=countCircle+1
    return countTotal, countCircle
square, circle= randomShots(10000000)
PI=4*(circle/square)
print(PI)
            
            
    
```
